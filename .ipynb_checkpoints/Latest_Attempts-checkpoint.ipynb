```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
import pandas as pd
import numpy as np
train = pd.read_csv("data/train_reduced.csv")
test = pd.read_csv("data/test_reduced.csv")

def feature_engineering(data):
    data['Dates'] = pd.to_datetime(data['Dates'], format="%m/%d/%y %H:%M")
    data['Day'] = data['Dates'].dt.day
    data['Month'] = data['Dates'].dt.month
    data['Year'] = data['Dates'].dt.year
    data['Hour'] = data['Dates'].dt.hour
    data['Minute'] = data['Dates'].dt.minute
    data['DayOfWeek'] = data['Dates'].dt.dayofweek
    data['WeekOfYear'] = data['Dates'].dt.weekofyear
    return data

X_train = feature_engineering(train)
X_test = feature_engineering(test)



X_train['street_corner'] = X_train['Address'].apply(lambda x: 1 if '/' in x else 0)
X_test['street_corner'] = X_test['Address'].apply(lambda x: 1 if '/' in x else 0)

X_train = X_train.drop(['Address', 'Category', 'Dates'], axis=1)
Y_train = train['Category']





X_test = X_test.drop(['Address', 'Dates'], axis=1)


```
